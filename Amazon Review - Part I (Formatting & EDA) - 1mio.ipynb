```python
#combining 10 different random files to reach 1 mio review

df = df0.append([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10])
df.index = np.arange(1, len(df)+1)
df.info()
```

```python
#Uniqueness of values
column_name=['marketplace',"customer_id","review_id","product_id","product_parent","star_rating","helpful_votes","total_votes","vine","verified_purchase","review_date"]
for i in column_name:
    print (f'"column:"{i}\nunique:"{df[i].unique().view()}\n')
```

```python
#CORRECTION OF COLUMNS AND TYPES
df = df.drop(df[(df.review_date == "1") | (df.review_date == "4") | (df.review_date == "2") | (df.review_date == "3") | (df.review_date == "5")| (df.review_date.isnull())].index)
df = df.drop(df[(df.star_rating == '2015-03-27')|(df.star_rating == "2015-04-17") | (df.star_rating == "2011-07-10") |  (df.star_rating.isnull())].index)
df['helpful_votes'] = df['helpful_votes'].fillna(0)
df['total_votes'] = df['total_votes'].fillna(0)
df= df.astype({"customer_id": int, "product_parent": int, "star_rating": int, "helpful_votes" : int, "total_votes":float, "review_body": str})
df['review_date']=pd.to_datetime(df['review_date'], format='%Y/%m/%d',errors ="ignore")

#adding review length and binary vectorize star rating by comment columns
df["review_len"]=[len(x) for x in df['review_body']]
df["comment"] = df['star_rating'].apply(lambda x: 1 if x>3 else 0)
```

# of rating by product

```python
new_df=df.groupby("product_id").filter(lambda x:x['star_rating'].count() >=300)
no_of_ratings_per_product = new_df.groupby(by='product_id')['star_rating'].count().sort_values(ascending=False)

fig = plt.figure(figsize=plt.figaspect(.5))
ax = plt.gca()
plt.plot(no_of_ratings_per_product.values)
plt.title('# rating per Product')
plt.xlabel('Product')
plt.ylabel('No of ratings per product')
ax.set_xticklabels([])

plt.show()
```

```python
print(f'product got most helpful review:\n {display(pd.DataFrame(no_of_ratings_per_product.head(5)))}')
```

# # MOST HELPFUL CUSTOMER

```python
new_df2=df.groupby("customer_id").filter(lambda x:x['helpful_votes'].count() >=50)
most_helpful_customers = new_df2.groupby(by='customer_id')['helpful_votes'].sum().sort_values(ascending=False)

fig = plt.figure(figsize=plt.figaspect(.2))
ax = plt.gca()
plt.plot(most_helpful_customers.values,)

plt.title('# review ')
plt.xlabel('customer',labelpad=True)
plt.ylabel('No of helpful votes')
ax.set_xticklabels([])

plt.show()
```

```python
print(f'most helpful customers: {display(pd.DataFrame(most_helpful_customers).head(5))}')
```
