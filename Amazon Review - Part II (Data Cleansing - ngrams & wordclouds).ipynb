# DATA CLEANSING

```python
sw = stopwords.words('english')
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from string import punctuation

def cleanText(text):
    #text = BeautifulSoup(text, "lxml",).text
    text = text.lower()

    text= text.replace('<br />',' ')
    text= text.replace('<br/>',' ')

    text = re.sub(r"won\'t", "will not", text)
    text = re.sub(r"can\'t", "can not", text)
    # general
    text = re.sub(r"n\'t", " not", text)
    text = re.sub(r"\'re", " are", text)
    text = re.sub(r"\'s", " is", text)
    text = re.sub(r"\'d", " would", text)
    text = re.sub(r"\'ll", " will", text)
    text = re.sub(r"\'t", " not", text)
    text = re.sub(r"\'ve", " have", text)
    text = re.sub(r"\'m", " am", text)
    text = re.sub(r'\|\|\| ', r'  ', text)
    text = re.sub(r'http\S+', r' <URL> ', text)
    text = re.sub(r"[-()\"#/@;:<>{}^`+=~|.!?,]", " ", text) 
    text = re.sub(r'[^a-zA-z.,!?/:;\"\'\s]', " ", text)
    text = re.sub(r'[^a-zA-z.,!?/:;\"\'\s]', " ", text)
    text = re.sub(r'[^\w]', ' ', text)
    text = re.sub("\s\s+", " ", text)
   # text=re.sub('(\\b[A-Ka-k] \\b|\\b [A-Ka-k]\\b)', '', text)
    return text
```

```python
df['cleaned_review'] = df['review_body'].apply(cleanText)
```

```python
def normalized_sentence(s):
    tokens = nltk.tokenize.word_tokenize(s)
    # remove short words, they're probably not useful
    tokens = [t for t in tokens if len(t) > 2]
    # put words into base form
    #tokens = [t for t in tokens if t not in stopwords.words("english") + [""]]
    return tokens
```

```python
df['tokenized_review'] = df['cleaned_review'].apply(normalized_sentence)
```

```python
df["tokenized_string"] = df.tokenized_review.apply(lambda x: ' '.join([str(i) for i in x]))
```
